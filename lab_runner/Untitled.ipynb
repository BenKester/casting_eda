```python
from data import Aggregated
from collections import defaultdict
```

```python
agg = Aggregated()
```

```python
uniques = defaultdict(lambda: {'count': 0, 'implicits': defaultdict(int)})
bases = defaultdict(int)
```

```python
for enchant in agg.values():
    if 'poe trade stats' in enchant:
        obj = enchant['poe trade stats']
        if obj != 'not found':
            for k, v in obj.get('uniques', {}).items():
                if v >= 2:
                    uniques[k]['count'] += v
                    for k_2, v_2 in obj.get('implicits', {}).items():
                        if v_2 >= 2:
                            uniques[k]['implicits'][k_2] += 1
            for k, v in obj.get('rare base types', {}).items():
                if v >= 2:
                    bases[k] += v
```

```python
{k:v for k, v in bases.items() if v > 5}
```
